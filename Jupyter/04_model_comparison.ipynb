# Qriton HLM — Model Comparison

Compare energy landscapes between two models. Find shared vs unique basins.
Transplant concepts from one model to another.

```python
import torch
from qriton_hlm import BasinSurgeon

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# Two synthetic models (replace with checkpoints)
d = 128
torch.manual_seed(42)
W_a = torch.randn(d, d, device=DEVICE) * 0.02
W_a = (W_a + W_a.T) / 2

torch.manual_seed(99)
W_b = torch.randn(d, d, device=DEVICE) * 0.02
W_b = (W_b + W_b.T) / 2

surgeon_a = BasinSurgeon.from_W(W_a, device=DEVICE)
surgeon_b = BasinSurgeon.from_W(W_b, device=DEVICE)
```

## Compare basin overlap

```python
diff = surgeon_a.compare(surgeon_b, layer=0)
print(f'Model A: {diff["basins_self"]} basins')
print(f'Model B: {diff["basins_other"]} basins')
print(f'Shared:  {diff["shared"]}')
print(f'Only A:  {diff["only_self"]}')
print(f'Only B:  {diff["only_other"]}')
```
