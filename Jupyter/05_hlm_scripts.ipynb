## Write a .hlm script file

```python
%%writefile make_polite.hlm
# Make a model more polite
load model.pt
set strength 0.1

# Capture examples
capture 5 polite Thank you so much
capture 5 polite I really appreciate it
capture 5 polite That's very kind of you

# Inject the concept
inject-concept 5 polite 0.1
apply 5

# Verify
benchmark
generate Tell me about the weather

# Cleanup
restore 5
```

```python
# Run it from CLI:
# !qriton-hlm --script make_polite.hlm

# Or from Python:
from qriton_hlm.cli import EnergyLang
lang = EnergyLang()
lang.run_script('make_polite.hlm')
```
